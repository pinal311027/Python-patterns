```python
#hollow square
n = 4
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print()
```

```python
#plus pattern
n = 5

for i in range(n):
    for j in range(n):
        if i==n//2 or j==n//2:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
#Pyramid
n = 4

for i in range(n):
    print(" "*(n-i-1), end="")
    print("*"*(2*i+1))
```

```python
#Pyramid
n = 4

for i in range(n):
    print(" "*i, end="")
    print("*"*(2*(n-i)-1))
```

```python
#hollow pyramid
n = 4

for i in range(n):
    for j in range(2*n-1):
        if j==n-i-1 or j==n+i-1 or i==n-1:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
#pascal's triangle
n = 5

for i in range(n):
    print(" "*(n-i), end="")
    num = 1
    for j in range(i+1):
        print(num, end=" ")
        num = num * (i-j) // (j+1)
    print()
```

```python
#binary triangle
n = 4

for i in range(1,n+1):
    for j in range(1,i+1):
        print((i+(j-1))%2,end=" ")
    print()
```

```python
#dimand
n = 4

for i in range(n):
    print(" "*(n-i-1), end="")
    print("*"*(2*i+1))

for i in range(n-2,-1,-1):
    print(" "*(n-i-1), end="")
    print("*"*(2*i+1))
```

```python
#hollow dimand
n = 4

for i in range(n):
    print(" "*(n-i-1), end="")
    if i==0:
        print("*")
    else:
        print("*"+" "*(2*i-1)+"*")

for i in range(n-2,-1,-1):
    print(" "*(n-i-1), end="")
    if i==0:
        print("*")
    else:
        print("*"+" "*(2*i-1)+"*")
```

```python
#butterfly pattern
n = 4

for i in range(1,n+1):
    print("*"*i+" "*(2*(n-i))+"*"*i)

for i in range(n-1,0,-1):
    print("*"*i+" "*(2*(n-i))+"*"*i)
```

```python
#x-pattern
n = 5

for i in range(n):
    for j in range(n):
        if i==j or i+j==n-1:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```
